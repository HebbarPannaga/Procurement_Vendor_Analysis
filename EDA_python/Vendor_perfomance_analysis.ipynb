```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3

sns.set_theme(style='whitegrid')
```

```python
conn = sqlite3.connect(r"..\inventory.db")
vendor_sales_summary_df = pd.read_sql_query("SELECT * FROM vendor_sales_summary",conn)
```

# Correlation analysis

```python
num_df = vendor_sales_summary_df.select_dtypes(include='number')
num_df.drop(columns=['VendorNumber', 'Brand'], inplace=True)
corr = num_df.corr()
plt.figure(figsize=(10,10))

sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.show()
```

# Top vendors by gross profit margin

```python
vendor_profit = vendor_sales_summary_df.groupby('VendorName', as_index=False).agg({'TotalSalesDollars':'sum', 'GrossProfit':'sum'})
vendor_profit = vendor_profit[vendor_profit['TotalSalesDollars'] > 0]
vendor_profit['Gross_profit_margin'] = vendor_profit['GrossProfit'] / vendor_profit['TotalSalesDollars']
vendor_profit = vendor_profit.sort_values(by='Gross_profit_margin', ascending=False).head(20)

plt.figure(figsize=(8,6))

ax = sns.barplot(data=vendor_profit,x='VendorName',y='Gross_profit_margin',color='steelblue')

plt.title('Top Vendors by Gross Profit Margin')
plt.xlabel('Vendor')
plt.ylabel('Gross profit margin')
plt.xticks(rotation=90)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
        (p.get_x() + p.get_width()/2, p.get_height()),
        ha='center',
        va='bottom', rotation=90)
plt.show()
```

# Vendor spend vs Gross profit

```python
vendor_profit = vendor_sales_summary_df.groupby('VendorName', as_index=False).agg(Spend=('TotalPurchaseDollars','sum'),GrossProfit=('GrossProfit','sum'))
plt.figure(figsize=(8,6))

plot_df = vendor_profit[vendor_profit['GrossProfit']< vendor_profit['GrossProfit'].quantile(0.95)]
sns.scatterplot(data=plot_df,x='Spend',y='GrossProfit',alpha=0.6, s=180)

plt.title("Vendor Spend vs Gross Profit")
plt.xlabel("Procurement Spend")
plt.ticklabel_format(style='plain', axis='y')
plt.ticklabel_format(style='plain', axis='x')
plt.ylabel("Gross Profit")

plt.show()
```

- Most vendors are concentrated within a similar procurement spend range, while a small number of vendors account for substantially higher procurement expenditure
- Gross profit generally increases with procurement spend, although the relationship is not perfectly proportional.

# Excise burden analysis

```python
excise = vendor_sales_summary_df[vendor_sales_summary_df['TotalSalesDollars'] > 0].copy()
excise['ExcisePercent'] = (excise['TotalExciseTax']/excise['TotalSalesDollars']) * 100
excise = excise.nlargest(15,'ExcisePercent')

plt.figure(figsize=(8,6))

sns.barplot(data=excise,y='Description',x='ExcisePercent',palette='flare')

plt.title("Products with Highest Excise Burden")
plt.show()
```

# Procurement ROI

```python
vendor_sales_summary_df['ProcurementROI'] = (vendor_sales_summary_df['GrossProfit']/vendor_sales_summary_df['TotalPurchaseDollars']) * 100
ROI = vendor_sales_summary_df.nlargest(15,'ProcurementROI')

plt.figure(figsize=(8,6))

sns.barplot(data=ROI,y='Description',x='ProcurementROI',palette='crest')

plt.title("Highest Procurement ROI")
plt.show()
```

- The highest-ranking products generate significantly more gross profit for every procurement dollar invested.
- These products represent the most efficient procurement investments and may be suitable candidates for prioritization

# High spend vs low margin

```python
import numpy as np
spend_margin = vendor_sales_summary_df[vendor_sales_summary_df['ProfitMargin'] != -np.inf].copy()
plt.figure(figsize=(10,7))

plot_df = spend_margin[(spend_margin['ProfitMargin'] > spend_margin['ProfitMargin'].quantile(0.1)) &
    (spend_margin['TotalPurchaseDollars'].between(spend_margin['TotalPurchaseDollars'].quantile(0.01),spend_margin['TotalPurchaseDollars'].quantile(0.95)))]

sns.scatterplot(data=plot_df,x='TotalPurchaseDollars',y='ProfitMargin',size='GrossProfit',sizes=(30,300), alpha=0.3)
plt.ticklabel_format(style='plain', axis='x')
plt.title("Purchase Spend vs Profit Margin")

plt.show()
```

- Higher procurement spend does not consistently translate into higher profit margins
- Several products with moderate procurement spend achieve relatively high margins, while some higher-spend products operate with comparatively lower margins

# Sales revenue vs Gross profit

```python
sales = vendor_sales_summary_df[vendor_sales_summary_df['TotalSalesDollars'] > 0].copy()

plt.figure(figsize=(8,6))

sns.scatterplot(data=sales,x='TotalSalesDollars',y='GrossProfit',size='TotalPurchaseQuantity',hue='ProfitMargin',sizes=(40,300),alpha=0.6)

plt.title("Sales Revenue vs Gross Profit")
plt.xlabel("Total Sales Revenue")
plt.ylabel("Gross Profit")
plt.ticklabel_format(style='plain', axis='y')
plt.xscale("log")
plt.show()
```

- Products with higher sales revenue generally generate higher gross profit
- However, products with similar sales revenue can produce noticeably different gross profits, suggesting variability in procurement costs or pricing

# Slow moving products

```python
slow = vendor_sales_summary_df[vendor_sales_summary_df['StockTurnover'] > 0].nsmallest(20,'StockTurnover')
slow = slow.sort_values(by='StockTurnover')
plt.figure(figsize=(8,6))

sns.barplot(data=slow,y='Description',x='StockTurnover',palette='magma')
plt.title("Slowest Moving Products")

plt.show()
```

- These products exhibit the lowest stock turnover, indicating slower inventory movement and a higher likelihood of inventory remaining in storage for extended periods.

# Sales Revenue efficiency

```python
top = vendor_sales_summary_df.nlargest(15,'SalestoPurchaseRatio')

plt.figure(figsize=(8,6))

sns.barplot(data=top,y='Description',x='SalestoPurchaseRatio',palette='viridis')

plt.title("Highest Revenue per Procurement Dollar")

plt.show()
```

# Vendor KPI

```python
vendor_score = (vendor_sales_summary_df.groupby('VendorName')
      .agg(
          Spend=('TotalPurchaseDollars','sum'),Revenue=('TotalSalesDollars','sum'),
          GrossProfit=('GrossProfit','sum'),Freight=('FreightCost','mean'),
          Excise=('TotalExciseTax','mean'),StockTurnover=('StockTurnover','mean'),
          SalesRatio=('SalestoPurchaseRatio','mean')
      )
      .sort_values('GrossProfit',ascending=False))
vendor_score
```

# Opportunity matrix

```python
plt.figure(figsize=(8,6))

plot_df = vendor_sales_summary_df[vendor_sales_summary_df['StockTurnover'].between(vendor_sales_summary_df['StockTurnover'].quantile(0.1),vendor_sales_summary_df['StockTurnover'].quantile(0.9))]
sns.scatterplot(data=plot_df,x='StockTurnover',y='GrossProfit',size='TotalSalesDollars',sizes=(40,300), alpha=0.3)

plt.title("Stock Turnover vs Gross Profit")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

- Products with higher stock turnover are distributed across a wide range of gross profits
- No strong relationship is observed between inventory turnover and gross profit

# Loss making products

```python
loss_products = vendor_sales_summary_df[vendor_sales_summary_df['ProfitMargin'] < 0]
loss_vendor = (loss_products.groupby('VendorName').agg(LossProducts=('Description','count'),Purchase=('TotalPurchaseDollars','sum'))
    .sort_values('LossProducts',ascending=False))
loss_vendor
```

```python
plot_df = loss_products[loss_products['ProfitMargin'] != -np.inf]
plot_df = plot_df[plot_df['ProfitMargin'].between(plot_df['ProfitMargin'].quantile(0.05), plot_df['ProfitMargin'].quantile(0.99))]
sns.histplot(plot_df['ProfitMargin'], bins=30)
plt.show()
```

- Most loss-making products are concentrated at moderately negative profit margins, indicating relatively limited losses for the majority of affected products
- The distribution includes a small number of extreme negative margin outliers, suggesting exceptional cases such as very low sales values or unusual transactions

# Capital locked in inslod inventory

```python
unsold = vendor_sales_summary_df[vendor_sales_summary_df['TotalSalesQuantity'] == 0]

print(f"Unsold Products : {len(unsold):,}")
print(f"Inventory Value : ${unsold['TotalPurchaseDollars'].sum():,.2f}")
```

# Which vendors contribute most to unsold inventory

```python
vendor_unsold = unsold.groupby('VendorName', as_index=False).agg(UnsoldValue=('TotalPurchaseDollars','sum')).sort_values('UnsoldValue',ascending=False).head(15)
plt.figure(figsize=(12,6))

sns.barplot(data=vendor_unsold,y='VendorName',x='UnsoldValue',palette='Reds_r')
plt.title("Top Vendors with Highest Unsold Inventory Value")

plt.xlabel("Unsold Inventory Value ($)")
plt.ylabel("Vendor")

plt.show()
```

# Unsold inventory %

```python
unsold_percent = (unsold['TotalPurchaseDollars'].sum()/vendor_sales_summary_df['TotalPurchaseDollars'].sum()) * 100
print(f"Unsold Inventory = {unsold_percent:.2f}%")
```

# Final insights

- Vendor profitability differs significantly even at similar procurement spend levels, indicating that purchase volume alone does not determine financial performance.
- Higher sales generally generate higher gross profit, although products with comparable sales revenue can exhibit different profitability due to pricing or procurement cost differences.
- Inventory turnover and profitability show limited correlation, suggesting that inventory performance should be evaluated alongside financial KPIs.
- Most loss-making products incur moderate negative margins, while extreme losses are limited to a small number of exceptional cases.
- Unsold inventory ties up procurement capital without contributing to revenue, highlighting opportunities for inventory optimization and demand planning.
- Vendor performance metrics such as procurement ROI, sales-to-purchase ratio, gross profit, and stock turnover support data-driven supplier evaluation and procurement decision-making.
